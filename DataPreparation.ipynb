```python
import numpy as np
from scipy.io.wavfile import read, write
from IPython.display import Audio
import matplotlib.pyplot as plt

MAX_SAMPLES = 7000
```

```python
def standardize_data(data):
    # if we are dealing with two-dimensions array, we want to consider only the first column data
    if (len(data.shape) > 1):
        data = data[:,0]

    # if the samples given exceed the MAX_SAMPLES constant, we are splitting it
    # otherwise, we want to add zeros until we reach a lenght equal to MAX_SAMPLES
    # Doing it, we will have always the same length for all the files given in input
    if(len(data)>MAX_SAMPLES):
        finalData = data[:MAX_SAMPLES]
    else:
        data = np.pad(data, (0, MAX_SAMPLES - len(data)), 'constant')
        finalData = data
        
    return finalData
```

```python
def plotData(data):
    plt.figure()
    plt.plot(data)
    plt.xlabel("Samples")
    plt.ylabel("Amplitude")
    plt.title("Wave")
    plt.show()
```

```python
# Read() function accept one-dimension or two-dimensions arrays
Fs, data = read('Kick_90s.wav')

data = standardize_data(data)

# Samples [n] = F_s [1/s] * seconds [s]
# Fs [Hz] * s [seconds] = 4800 * 11 = 500000 samples 
print("Sampling frequency is", Fs)


plotData(data)

print("Listen to the audio!")
Audio(data, rate = Fs)
```
